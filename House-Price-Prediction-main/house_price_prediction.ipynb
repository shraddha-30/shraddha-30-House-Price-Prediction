```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

# 2- load  train.csv of the house-price-prediction-challenge dataset

```python
dataframe_train=pd.read_csv("/content/Dataset/train.csv")
dataframe_train
```

# 3- load  test.csv of the house-price-prediction-challenge dataset

```python
dataframe_test=pd.read_csv("/content/Dataset/test.csv")
dataframe_test
```

```python
# Drop 'ADDRESS' column from both datasets
dataframe_train.drop(columns=['ADDRESS'], inplace=True)
dataframe_test.drop(columns=['ADDRESS'], inplace=True)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns (adjust as per your dataset structure)
categorical_cols = ['POSTED_BY', 'BHK_OR_RK']

# Initialize LabelEncoder
label_encoders = {}

# Encode categorical columns for train dataset
for col in categorical_cols:
    le = LabelEncoder()
    dataframe_train[col] = le.fit_transform(dataframe_train[col])
    label_encoders[col] = le  # Save the encoder for future use

# Encode categorical columns for test dataset using existing encoders
for col in categorical_cols:
    le = label_encoders[col]
    dataframe_test[col] = le.transform(dataframe_test[col])

# Display the preprocessed datasets
print("Train Dataset:")
print(dataframe_train.head())
print("\nTest Dataset:")
print(dataframe_test.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Assuming dataframe_train and dataframe_test are already prepared and encoded

# Extracting features and target from dataframe_train
X_train = dataframe_train.drop(['TARGET(PRICE_IN_LACS)'], axis=1)
y_train = dataframe_train['TARGET(PRICE_IN_LACS)']

# Extracting features from dataframe_test
X_test = dataframe_test

# Initialize Linear Regression model
lin_reg_model = LinearRegression()

# Fit the model
lin_reg_model.fit(X_train, y_train)

# Predicting on test data
predictions = lin_reg_model.predict(X_test)

# Print or return predictions as needed
print("Predictions for test data:")
print(predictions)
```

```python
# Predicting on training data
y_train_pred = lin_reg_model.predict(X_train)

# Calculate MSE for training and test data
train_mse = mean_squared_error(y_train, y_train_pred)

# Print MSE for training and test data
print(f"Training MSE: {train_mse}")
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_train, y_train_pred)
print(f"Training R2 Score: {r2}")
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_train, y_train_pred)
print(f"Training MAE: {mae}")
```

```python
rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
print(f"Training RMSE: {rmse}")
```

```python
random_forest_model=RandomForestRegressor(n_estimators=200, random_state=7)
random_forest_model.fit(X_train,y_train)

y_prediction_random_forest=random_forest_model.predict(X_train)

y_prediction_random_forest_test=random_forest_model.predict(X_test)


print("RMSE:",np.sqrt(mean_squared_error(y_train,y_prediction_random_forest)))
print("R_square:",r2_score(y_train,y_prediction_random_forest))
```
